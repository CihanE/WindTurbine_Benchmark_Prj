```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#read_wind_turbine_data
Wt = pd.read_csv('C:/Users/ghane/Documents/WindTurbineProject/supply__wind_turbine_library.csv')
display(Wt.head(10))
display(Wt.info())
```

```python
Wt.describe(include='all')
Wt.isna().sum()
```

```python
Wt= Wt[~Wt['manufacturer'].isin(['aerodyn', 'IEA'])]
sns.countplot(Wt['manufacturer'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.stripplot(data=Wt, x='manufacturer', y='nominal_power')
plt.xticks(rotation=90)
plt.show()
```

```python
Wt['efficiency'] = Wt['nominal_power'] / Wt['rotor_area']
sns.relplot(data=Wt, x='nominal_power', y='efficiency', hue='manufacturer')
plt.show()
```

```python
sns.histplot(data=Wt, x='nominal_power', binwidth=200)
```

```python
group_names =['0-3k', '3k-5k', '5k-10k']
ranges = [0,3000,5000,np.inf]
Wt['power_cat'] = pd.cut(Wt['nominal_power'], bins=ranges, labels=group_names)
sns.countplot(data=Wt, x= 'power_cat').set(title='Power_by_Categories')
plt.show()
```

```python
sns.stripplot(data=Wt, x='manufacturer', y='efficiency', hue='power_cat')
plt.xticks(rotation=90)
plt.show()
```
